```python
dataset=pd.concat([dataset,Gender,Education,Property_Area],axis=1)
```

```python
dataset=dataset.drop(['Gender',"Education","Property_Area"], axis=1)
```

```python
dataset.loc[dataset['Married']=="Yes", 'Married']=1
dataset.loc[dataset['Married']=="No", 'Married']=0

dataset.loc[dataset['Self_Employed']=="Yes", 'Self_Employed']=1
dataset.loc[dataset['Self_Employed']=="No", 'Self_Employed']=0

dataset.loc[dataset['Loan_Status']=="Y", 'Loan_Status']=1
dataset.loc[dataset['Loan_Status']=="N", 'Loan_Status']=0

dataset.loc[dataset['Male']=="True", 'Male']=1
dataset.loc[dataset['Male']=="False", 'Male']=0

dataset.loc[dataset['Not Graduate']=="True", 'Not Graduate']=1
dataset.loc[dataset['Not Graduate']=="False", 'Not Graduate']=0

dataset.loc[dataset['Rural']=="True", 'Rural']=1
dataset.loc[dataset['Rural']=="False", 'Rural']=0

dataset.loc[dataset['Urban']=="True", 'Urban']=1
dataset.loc[dataset['Urban']=="False", 'Urban']=0

dataset.loc[dataset['Semiurban']=="True", 'Semiurban']=1
dataset.loc[dataset['Semiurban']=="False", 'Semiurban']=0
```

```python
dataset.head()
dataset.loc[dataset['Dependents']=="3+", 'Dependents']=3
dataset.reset_index()
```

```python
dataset['Loan_Status'] = dataset['Loan_Status'].astype(int) #1.0 0.0 bhanera set bhayo, binary classification ko final class 
#float ma cha bhane predict garna sakdaina. Integer ma lagna parcha
corr_matrix=dataset.corr()
corr_matrix["Loan_Status"].sort_values(ascending=False)
```

```python
from sklearn.model_selection import train_test_split
train_set, test_set=train_test_split(dataset, test_size=0.2, random_state=42)
```

```python
train_set['Loan_Status'].value_counts()
```

```python
test_set['Loan_Status'].value_counts()
```

```python
X=train_set.drop(['Loan_Status'],axis=1)
Y_train=train_set['Loan_Status']
Xt=test_set.drop(['Loan_Status'],axis=1)
Y_test=test_set['Loan_Status']
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
my_pipeline=Pipeline([('imputer',SimpleImputer(strategy='median')),('std-scaler',StandardScaler())])
```

```python
X_train=my_pipeline.fit_transform(X)
X_test=my_pipeline.fit_transform(Xt)
```

```python
Y_train.value_counts()
X_train.shape, Y_train.shape
dataset['Dependents'] = dataset['Dependents'].astype(int)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(X_train,Y_train)
```

```python
from sklearn.metrics import mean_squared_error
predictions=model.predict(X_train)
score=mean_squared_error(Y_train, predictions)
print(np.sqrt(score))
```

```python
from sklearn.model_selection import cross_val_score
score=cross_val_score(model, X_train, Y_train, scoring="accuracy", cv=100)
print(score.mean())
```

```python
from joblib import dump,load
dump(model, "LoanPrediction.joblib")
```
